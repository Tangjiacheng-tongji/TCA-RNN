```python
import tensortools as tt
import numpy as np
import matplotlib.pyplot as plt
```

```python
test_result=np.loadtxt("temp.txt").reshape((400,49,100))#750 trailsï¼Œ100 neurons and 49 time points
```

```python
A=tt.Ensemble(fit_method='cp_als')
```

```python
A.fit(test_result,ranks=range(1,6),replicates=3)
```

```python
plt.figure()
tt.plot_objective(A)
plt.show()
```

```python
plt.figure()
tt.plot_similarity(A)
plt.show()
```
